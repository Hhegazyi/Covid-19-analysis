```python
import numpy as np
from numpy import *
import math
import statistics
from fractions import Fraction
from scipy.stats import linregress
```

```python
#Constants:
PI=math.pi
E=math.e

#Memory storage for operations:
memory_store=0


#arethmatic operations:
def add(x,y):
    return x+y
def sub(x,y):
    return x-y
def multiply(x,y):
    return x*y
def divide(x,y):
    return x/y

#advanced functions:
def powerFunc(x,y):
    return x**y
def sqrtFunc(x):
    return math.sqrt(x)
def nthrootFunc(x,y):
    return x**(1/y)

#trigonometric functions:
def sinFunc(x):
    return np.sin(np.radians(x))
def cosFunc(x):
    return np.cos(np.radians(x))
def tanFunc(x):
    return np.tan(np.radians(x))

#Hyperbolic function:
def sinhFunc(x):
    return np.sinh(np.radians(x))
def coshFunc(x):
    return np.cosh(np.radians(x))
def tanhFunc(x):
    return np.tanh(np.radians(x))

#Logarithmic functions:
def lnFunc(x):
    return np.log(x)
def log10Func(x):
    return np.log10(x)
def log2Func(x):
    return np.log2(x)
def expFunc(x):
    return np.exp(x)

#statistical functions:
def meanFunc(data):
    return np.mean(data)
def stdDev(data):
    return np.std(data)
def variance(data):
    return np.var(data)
def regression(x,y):
    slope,intercept,r_value,p_value,std_err=linregress(x,y)
    return slope, intercept,r_value **2

#Fractional and Integer Functions:
def factorial(n):
    if n < 0:
        return "Math error (factorial of negative number)"
    return math.factorial(n)
def combination(n,r):
    return math.comb(n,r)
def permutation(n,r):
    return math.perm(n,r)
def fractionToDecimal(nem,den):
    return float(Fraction(num,den))
def decimalToFraction(decimal):
    return Fraction(decimal).limit_denominator()

#Memory Functions:
def memoryStore(value):
    global memory_store
    memory_store = value
def memoryRecall():
    return memory_store
def memoryAdd(value):
    global memory_store
    memory_store += value
def memorySubtract(value):
    global memory_store
    memory_store -= value
def memoryClear():
    global memory_store
    memory_store = 0

#Matrix operations:
def matrixAdd(A, B):
    return np.add(A, B)
def matrixSubtract(A, B):
    return np.subtract(A, B)
def matrixMultiply(A, B):
    return np.dot(A, B)
def matrixDeterminant(A):
    return np.linalg.det(A)
def matrixInverse(A):
    if np.linalg.det(A) == 0:
        return "Matrix is singular"
    return np.linalg.inv(A)

# Constants:
def getConstant(constant):
    if constant == "pi":
        return PI
    elif constant == "e":
        return E
    else:
        return "Invalid constant"

#the operation of the calculator:
print("\nSelect an option:")
print("1. Basic Arithmetic")
print("2. Advanced Math")
print("3. Trigonometry")
print("4. Hyperbolic Functions")
print("5. Logarithms")
print("6. Statistical Functions")
print("7. Fractional and Integer Functions")
print("8. Memory Functions")
print("9. Matrix Operations")
print("10. Constants")
choice = input("Enter choice: ")

#arethmatic:
if choice=="1":
    n1=float(input("enter the first number"))
    op=input("enter the operation(+,-,x,/)")
    n2=float(input("enter the second number"))

    if op=="+":
        print(n1,op,n2,"=",add(n1,n2))
    elif op=="-":
        print(n1,op,n2,"=",sub(n1,n2))
    elif op=="x":
        print(n1,op,n2,"=",multiply(n1,n2))
    elif op=="/":
        if n2!=0:
           print(n1,op,n2,"=",divide(n1,n2))
        else:
           print(n1,op,n2,"=","Math error")
    else:
        print("invalid operation")

#advanced math:
elif choice=="2":
    n1=float(input("enter the base no."))
    op=input("Enter the operation (power, sqrt, nthroot)")
    if op=="power":
        n2=float(input("enter the power no."))
        print(n1,"^",n2,"=",powerFunc(n1,n2))
    elif op=="sqrt":
        print("the square root of",n1,"=",sqrtFunc(n1))
    elif op=="nthroot":
        n2=float(input("enter the root degree"))
        print("the",n2,"root of",n1,"=",nthrootFunc(n1,n2))

#Trigonometric:
elif choice=="3":
    Trigfunc=input("enter the trig function(sin,cos,tan)")
    n=float(input("enter the angle"))
    if Trigfunc=="sin":
        print(Trigfunc,"(",n,")","=",sinFunc(n))
    elif Trigfunc=="cos":
        print(Trigfunc,"(",n,")","=",cosFunc(n))
    elif Trigfunc=="tan":
        print(Trigfunc,"(",n,")","=",tanFunc(n))
    else:
        print("invalid operation")

#Hyperbolic
elif choice=="4":
    Hypfunc=input("enter the hyperbolic function(sinh,cosh,tanh)")
    c=float(input("enter the angle"))
    if Hypfunc=="sinh":
        print(Hypfunc,"(",c,")","=",sinhFunc(c))
    elif Hypfunc=="cosh":
        print(Hypfunc,"(",c,")","=",coshFunc(c))
    elif Hypfunc=="tanh":
        print(Hypfunc,"(",c,")","=",tanhFunc(c))
    else:
        print("invalid operation")

#Logarithmic:
elif choice=="5":
    op=input("Enter the operation:(Ln, Log10, Log2, exp.)")
    if op=="ln":
        n=float(input("enter the no."))
        print(op,"(",n,")","=",lnFunc(n))
    elif op=="log10":
        n=float(input("enter the no."))
        print(op,"(",n,")","=",log10Func(n))
    elif op=="log2":
        n=float(input("enter the no."))
        print(op,"(",n,")","=",log2Func(n))
    elif op=="exp.":
        n=float(input("enter the power of e"))
        print(op,"(",n,")","=",expFunc(n))
    else:
        print("invalid operation")

# Statistical functions
elif choice == "6":
    stat_func = input("Enter the statistical function (mean, stddev, variance, regression): ")
    data = list(map(float, input("Enter a list of numbers separated by spaces: ").split()))

    if stat_func == "mean":
        print(f"Mean = {meanFunc(data)}")
    elif stat_func == "stddev":
        print(f"Standard Deviation = {stdDev(data)}")
    elif stat_func == "variance":
        print(f"Variance = {variance(data)}")
    elif stat_func == "regression":
        y = list(map(float, input("Enter another list of numbers for regression separated by spaces: ").split()))
        slope, intercept, r_squared = regression(data, y)
        print(f"Regression: y = {slope}x + {intercept}, R^2 = {r_squared}")
    else:
        print("Invalid operation")


# Fractional and integer functions
elif choice == "7":
    frac_func = input("Enter the fractional or integer function (factorial, nCr, nPr, fraction2decimal, decimal2fraction): ")

    if frac_func == "factorial":
        n = int(input("Enter a number: "))
        print(f"{n}! = {factorial(n)}")
    elif frac_func == "nCr":
        n = int(input("Enter n: "))
        r = int(input("Enter r: "))
        print(f"{n} choose {r} = {combination(n, r)}")
    elif frac_func == "nPr":
        n = int(input("Enter n: "))
        r = int(input("Enter r: "))
        print(f"{n} permute {r} = {permutation(n, r)}")
    elif frac_func == "fraction2decimal":
        numerator = int(input("Enter the numerator: "))
        denominator = int(input("Enter the denominator: "))
        print(f"Decimal of {numerator}/{denominator} = {fractionToDecimal(numerator, denominator)}")
    elif frac_func == "decimal2fraction":
        decimal = float(input("Enter the decimal: "))
        print(f"Fraction of {decimal} = {decimalToFraction(decimal)}")
    else:
        print("Invalid operation")

# Memory functions
elif choice == "8":
    memory_func = input("Enter the memory function (store, recall, add, subtract, clear): ")

    if memory_func == "store":
        value = float(input("Enter value to store: "))
        memoryStore(value)
        print(f"Stored {value} in memory.")
    elif memory_func == "recall":
        print(f"Recalled value from memory: {memoryRecall()}")
    elif memory_func == "add":
        value = float(input("Enter value to add to memory: "))
        memoryAdd(value)
        print(f"Added {value} to memory. Current memory: {memoryRecall()}")
    elif memory_func == "subtract":
        value = float(input("Enter value to subtract from memory: "))
        memorySubtract(value)
        print(f"Subtracted {value} from memory. Current memory: {memoryRecall()}")
    elif memory_func == "clear":
        memoryClear()
        print("Memory cleared.")
    else:
        print("Invalid operation")


# Matrix operations
elif choice == "9":
    matrix_op = input("Enter the matrix operation (add, subtract, multiply, determinant, inverse): ")

    if matrix_op in ["add", "subtract", "multiply"]:
        A = np.array(eval(input("Enter matrix A in the form [[a11, a12], [a21, a22], ...]: ")))
        B = np.array(eval(input("Enter matrix B in the form [[b11, b12], [b21, b22], ...]: ")))

        if matrix_op == "add":
            print(f"Matrix A + Matrix B = \n{matrixAdd(A, B)}")
        elif matrix_op == "subtract":
            print(f"Matrix A - Matrix B = \n{matrixSubtract(A, B)}")
        elif matrix_op == "multiply":
            print(f"Matrix A * Matrix B = \n{matrixMultiply(A, B)}")

    elif matrix_op == "determinant":
        A = np.array(eval(input("Enter matrix A in the form [[a11, a12], [a21, a22], ...]: ")))
        print(f"Determinant of matrix A = {matrixDeterminant(A)}")
    elif matrix_op == "inverse":
        A = np.array(eval(input("Enter matrix A in the form [[a11, a12], [a21, a22], ...]: ")))
        print(f"Inverse of matrix A = \n{matrixInverse(A)}")
    else:
        print("Invalid operation")

# Constants
elif choice == "10":
    const = input("Enter the constant (pi, e): ")
    print(const,'=',getConstant(const))


else:
    print("Invalid choice")
```
